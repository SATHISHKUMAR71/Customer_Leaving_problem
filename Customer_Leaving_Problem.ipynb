#importing libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf

tf.__version__
```

# Data Preprocessing

```python
dataset = pd.read_csv("Churn_Modelling.csv")
x = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
print(y)
```

# Encoding Categorical Data

Encoding Gender

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
```

```python
print(x)
```

one hot encoding for Geography

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#ANN

```python
Artificial_NN = tf.keras.models.Sequential()
```

```python
Artificial_NN.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
Artificial_NN.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
Artificial_NN.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

# Compiling the NN

```python
Artificial_NN.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Training the Aritificial_NN

```python
Artificial_NN.fit(X_train, y_train, batch_size = 32, epochs = 400)
```

#Predict

```python
print(Artificial_NN.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]]))>0.5)
```
